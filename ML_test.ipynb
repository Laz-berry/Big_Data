```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
display(x_test.head())

customerid = x_test['CustomerId']
```

```python
def get_var_type(dataframe):
    continuous_var = []
    categorical_var = []
    
    for columns in dataframe.columns:
        if dataframe[columns].dtype == 'object':
            categorical_var.append(columns)
        else:
            continuous_var.append(columns)
    return continuous_var, categorical_var
```

```python
continueous_vars, categorical_vars = get_var_type(x_train)

print(continueous_vars)
print(categorical_vars)
```

```python
continueous_vars.remove('CustomerId')
categorical_vars.remove('Surname')
```

```python
x_train = pd.get_dummies(x_train, columns=categorical_vars)
x_test = pd.get_dummies(x_test, columns=categorical_vars)

x_train
```

```python
x_train.drop(['CustomerId', 'Surname'], axis=1, inplace=True)
x_test.drop(['CustomerId', 'Surname'], axis=1, inplace=True)

x_train
```

```python
from sklearn.model_selection import train_test_split

data = x_train
target = y_train['Exited']

train_X, test_X, train_y, test_y = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
print(train_X.shape)
print(test_X.shape)
print(train_y.shape)
print(test_y.shape)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, f1_score

LGB_model = LGBMClassifier()

LGB_model.fit(train_X, train_y)
train_pred = LGB_model.predict(test_X)

print('train score :', accuracy_score(test_y, train_pred))
```

```python
test_pred = LGB_model.predict(x_test)
```
